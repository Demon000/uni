```python
t=var('t')
x=function('x')(t)
w0=var('w0')
assume(w0 > 0)
x0=var('x0')
v0=var('v0')
eq=diff(x, 2) + w0^2 * x == 0
sol=desolve(eq, [x, t])
show(sol)
```

```python
cond=[0, x0, v0]
sol_cond=desolve(eq, [x, t], ics=cond)
show(sol_cond)
```

```python
plot(sol_cond.substitute(x0=1, w0=1, v0=1), [0, 4 * pi / 1])
```

```python
forget()
t=var('t')
la=var('la')
x=function('x')(t)
w0=var('w0')
assume(la^2 > 4 * w0)
x0=var('x0')
eq=diff(x, 2) + la * diff(x, 1) + w0 * x == 0
sol=desolve(eq, [x, t])
show(sol)
```

```python
sol_cond=desolve(eq, [x, t], ics=[0, 1, 5])
show(sol_cond)
```

```python
plot(sol_cond.substitute(la=25, w0=10), [0, 5])
```
